Ejercicio 3 - Aplicación IA: aproximación de gradiante

```python
import numpy as np # Biblioteca NumPy

def loss(w): # Define función de perdida
    return w**2 + 3*w + 1 # Define el polinomio

w = 1.5 # Valor del parámetro del modelo

h = 0.0001 # Paso pequeño

grad = (loss(w+h)-loss(w-h))/(2*h) # Apriximación del gradiante

real = 2*w + 3 # Gradiante real

print("Gradiante numérico:", grad) # Muestra gradiante aproximado

print("Gradiante real:", real) # Muestra gradiante exacto
```
