Ejercicio 1 - Regla del trapecio

```python
import numpy as np # Importa la biblioteca para cáculos numéricos

def f(x): #Define la función a integrar
    return x**2 # Retorna el valor de la función x^2

a = 0 # Límite inferior de integración
b = 2 # Límite superior de integración
n = 1000 # Número de subintervalos

h = (b-a)/n # Calcula el tamaño del paso

x = np.linspace(a,b,n+1) # Genera los puntos de discretización

y = f(x) # Evalúa la función en todos los puntos

integral = (h/2)*(y[0] + 2*np.sum(y[1:n]) + y[n]) # Aplica la fórmula del trapecio compuesto

print("Integral aproximada (Trapecio):",integral) # Muestra el resultado numérico
```
