Ejercicio 2 - Diferencia central

```python
import numpy as np # Importa NumPy

def f(x): # Define función matemática
    return x**3

x = 2 # Punto de evaluación

h = 0.001 # Paso pequeño

derivada = (f(x+h)-f(x-h))/(2*h) # Fórmula de diferencia central

real = 3*x**2 # Derivada exacta

error = abs(real-derivada) # Error absoluto

print("Derivada central:", derivada) # Resultado numérico

print("Derivada real:", real) # Resultado exacto

print("Error:", error) # Diferencia entre ambos
```
