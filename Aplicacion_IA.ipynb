Ejercicio 3 - Apliacación IA, integración tipo probabilidad

```python
import numpy as np  # Importa NumPy

def f(x,y):        # Define función tipo gaussiana
    return np.exp(-(x**2 + y**2))  # Calcula e^(-(x^2+y^2))

a, b = -1, 1  # Intervalo en x
c, d = -1, 1  # Intervalo en y

n = 100  # Número de divisiones en x
m = 100  # Número de divisiones en y

dx = (b-a)/n  # Paso en x
dy = (d-c)/m  # Paso en y

integral = 0  # Inicializa acumulador

for i in range(n):  # Itera en x
    for j in range(m):  # Itera en y

        x = a + i*dx    # Calcula punto en x
        y = c + j*dy    # Calcula punto en y

        integral += f(x,y)*dx*dy  # Acumula contribución

print("Probabilidad aproximada:", integral)  # Muestra resultado
```
