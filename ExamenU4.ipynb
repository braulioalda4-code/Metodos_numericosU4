```python
import numpy as np # Importa la biblioteca NumPy para cálculos numéricos

def f(x): # Define la función matemática a analizar
    return  3*x**2 - 12*x + 9 # Retorna el valor de la fución x^3

x = 2 # Punto donde se calculará la derivada

h = 0.001 # Tamaño del paso numérrico

derivada = (f(x+h)-f(x))/h # Aplica la fórmula de diferencia hacia adelante

real = 3*x**2 - 12*x + 9 # Calcula la derivada analítica real

error = abs(real-derivada) # Calcula el error absoluto

print("Derivada numérica:", derivada) # Muestra el valor aproximado

print("Derivada real:", real) # Muestra el valor exacto

print("Error:", error) # Muestra la diferencia entre valores
```

```python
# PARTE B - Integración numérica de L(x) en [0,3]
import numpy as np  # Importa NumPy

def L(x):  # Define la función de pérdida
    return x**3 - 6*x**2 + 9*x + 1  # Función del problema

a, b = 0, 3  # Límites de integración
n = 100  # Número de subintervalos (debe ser par para Simpson)

# MÉTODO DEL TRAPECIO
h = (b-a)/n  # Calcula el tamaño del paso
x = np.linspace(a, b, n+1)  # Genera los nodos
y = L(x)  # Evalúa la función

# Aplica la fórmula del trapecio compuesto
trapecio = (h/2)*(y[0] + 2*np.sum(y[1:n]) + y[n])

# MÉTODO DE SIMPSON (requiere n par)
if n % 2 == 0:  # Verifica que n sea par
    suma = y[0] + y[n]  # Inicializa con extremos

    for i in range(1, n):  # Itera sobre puntos interiores
        if i % 2 == 0:  # Índice par
            suma += 2*y[i]  # Coeficiente 2
        else:  # Índice impar
            suma += 4*y[i]  # Coeficiente 4

    simpson = (h/3)*suma  # Fórmula de Simpson
else:
    simpson = None
    print("Advertencia: Simpson requiere n par")

# Integral exacta (cálculo analítico)
def F(x):  # Antiderivada
    return x**4/4 - 2*x**3 + (9/2)*x**2 + x

exacto = F(b) - F(a)

# Muestra resultados
print("="*50)
print("PARTE B - Integración numérica")
print("="*50)
print("Integral exacta:", exacto)
print("Método del trapecio:", trapecio)
print("Error del trapecio:", abs(trapecio - exacto))
print("Método de Simpson:", simpson)
print("Error de Simpson:", abs(simpson - exacto))
```
